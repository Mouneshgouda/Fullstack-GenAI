```python
import matplotlib.pyplot as plt
x=[2018, 2019, 2020]
y=[100, 150, 130]
plt.plot(x,y)
plt.title("Line Plot");
plt.xlabel("Year")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 3, 45, 6]
y = [4, 6, 7, 8]

plt.hist(x, bins=5)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 3, 45, 6]
y = [4, 6, 7, 8]

plt.bar(x,y)  # bar chart of x values
plt.show()
```
